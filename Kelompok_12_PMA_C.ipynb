# Praproses Data

```python
# Cek apakah terdapat baris yang kosong
tgp.isnull().sum()
```

```python
# Cek apakah terdapat baris yang kosong
tgp[tgp.isnull().any(axis=1)]
```

Pembuangan data menggunakan z-score


```python
tgp_mean = tgp.balance.mean()
tgp_stdev = tgp.balance.std()
print("Balance rata-rata: {0}\nBalance Standard Deviasi: {1}".format(tgp_mean, tgp_stdev))
```

```python
# define our treshold value for z-score
threshold = 1 # this means: we only want to tolerate values within ** 1 std. ** from the central value (mean)
```

```python
outlier = []
for i in tgp['balance']:
    z = (i-tgp_mean)/tgp_stdev
    if abs(z) < threshold:
        outlier.append(i)
print('Our outlier value in tgp is', outlier)
```

```python
from scipy import stats
tgp['z_score']= stats.zscore(tgp['balance'])
tgp
```

```python
tgp1 = tgp.copy()
tgp1 = tgp1.loc[tgp1['z_score'].abs() < threshold]
tgp1 = tgp1.reset_index(drop = True)
tgp1
```

Pembuangan data menggunakan Inter Quartile Range

```python
Q1 = tgp1['balance'].quantile(0.25)
Q3 = tgp1['balance'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
tgp1 = tgp1[(tgp1['balance'] >= lower_bound) & (tgp1['balance'] <= upper_bound)]
```

```python
Q1 = tgp1['age'].quantile(0.25)
Q3 = tgp1['age'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
tgp1 = tgp1[(tgp1['age'] >= lower_bound) & (tgp1['age'] <= upper_bound)]
```

```python
Q1 = tgp1['duration'].quantile(0.25)
Q3 = tgp1['duration'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
tgp1 = tgp1[(tgp1['duration'] >= lower_bound) & (tgp1['duration'] <= upper_bound)]
```

```python
numerical_features = ['age', 'balance', 'campaign', 'pdays', 'previous']
for feature in numerical_features:
    median_value = tgp1[feature].median()
    tgp1[feature].fillna(median_value, inplace=True)
```
